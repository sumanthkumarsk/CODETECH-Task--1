Gender Distrubtion

```python
plt.title("Gender distrubtion")
ax=sns.countplot(data,x="Gender",hue="WklyStudyHours")
plt.figure(figsize=(5,5))
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[2])
plt.show()
```

From above chart we analysed that the no of females in data is more that no of males

```python
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
for column in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(data[column], kde=True, bins=20, color='blue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
if len(numerical_columns) > 1:
    print("\nPairwise relationships:")
    sns.pairplot(data[numerical_columns], diag_kind='kde', markers='+')
    plt.show()
```

```python
if len(numerical_columns) > 1:
    plt.figure()
    plt.scatter(data[numerical_columns[0]], data[numerical_columns[1]], color='green', alpha=0.5)
    plt.title(f'Scatter Plot: {numerical_columns[0]} vs {numerical_columns[1]}')
    plt.xlabel(numerical_columns[0])
    plt.ylabel(numerical_columns[1])
    plt.show()
```

```python
groupdata=data.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(groupdata)
```

HeatMap

```python
sns.heatmap(groupdata,annot=True,cmap="coolwarm")
plt.title("HeatMap")
plt.show()
```

From above heatmap we can conclude that the education of parents have good impact on student scores

```python
matrialstatus=data.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
print(matrialstatus)
```

```python
sns.heatmap(matrialstatus,annot=True,cmap="coolwarm")
plt.title("HeatMap")
plt.show()
```

From above heatmap we can conclude that thre is no impact on the students score due to parent matrial status
